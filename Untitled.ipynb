The data set includes two csv files - a player data file and session data file. Here is a summary of information provided by each data frame: 

## players.csv

- The data frame contains **196 observations**, each representing a unique player.
- There are 7 columns, containing the following variables:
    - `experience`: a categorical variable (stored as a chr) with each value representing the extent of familiarity with Minecraft.
    - `susbscribe`: a logical variable representing whether a player is subscribed to the newsletter.
    - `hashedEmail`: a unique character containing the hased email attributed to a player.
    - `player_hours`: amount of hours played by a player (dbl).
    - `name`: name of a player (chr).
    - `gender`: gender of a player (chr).
    - `Age`: age of a player (dbl).
- I see a potential issue with the `experience` parameter as it is self-reported and may be innacurate unless there are clear guidelines that determine experience level. It is an important variable for evaluating how different players approaching the game, so innacuracies may constrain the research.
- Later on, it would be best to convert the `experience` variable into a factor, as it is categorical.

## sessions.csv

- The data frame contains **1535 observations**, each representing a session played on the Minecraft server.
- There are 5 columns, containing the following variables:
    - `hashedEmail`: a unique character containing the hased email attributed to a player.
    - `start_time`: a date and time variable (stored as a chr) representing the start time of a session.
    - `end_time`: a date and time variable (stored as a chr) representing the end time of a session.
    - `original_start_time`: start time of a session in the Unix epoch time format (stored as a dbl).
    - `original_end_time`: end time of a session in the Unix epoch time format (stored as a dbl).
- The `start_time` and `end_time` variables will need to be converted from a string into a date time variable to conduct analysis.
- I don't see a particualar use for the `original_start_time` and `original_end_time` variables as the format doesn't allow to calculate accurate time differences at small time frames (e.g. a 10 minute difference would be calculated as 0 if we use the original variables for calculation).


## Summary statistics:

Looking at the `players.csv`, let's first derive the number of players (total rows) and active players (non-zero played_hrs value):

```r
p_num <- nrow(player_data)
act_p_num <- player_data |>
    filter(played_hours != 0) |>
    nrow()

p_num
act_p_num
```

Running the cell above demonstrates that there are **85 players** who haven't contributed any playing time.

Now let's look at a statistical summary for time played to see the total, max, min, median and mean values.

```r
playtime_stat <- player_data |>
    summarize(total_playtime = sum(played_hours), 
              avg_playtime = mean(played_hours, na.rm = TRUE), 
              median_playtime = median(played_hours, na.rm = TRUE), 
              max_playtime = max(played_hours, na.rm = TRUE), 
              min_playtime = min(played_hours, na.rm = TRUE)) |>
    map_dfr(~ round(.x, 2))
playtime_stat
```

It is interesting to see a very low **median** playtime score of **0.1 hours**, while the **average** sits at nearly **6 hours**, showing that there are a lot of inactive data points that skew the data.

Additionally, it is noteworthy to look at a player breakdown by experience level.

```r
p_by_exp <- player_data |>
    group_by(experience) |>
    summarize(player_share = n()/p_num*100) |>
    mutate(player_share = round(player_share, 2))

p_by_exp
```

Over half of the players are linked to a level of experience above regular. However, it is uncertain how the experience level was determined.

Looking at more general information, below is a summary of age statistics, player breakdown by gender and the newsletter subscription rate.

```r
gender_pct <- player_data |>
    group_by(gender) |>
    summarize(player_share = n()/p_num*100) |>
    mutate(player_share = round(player_share, 2))

subscribed_pct <- player_data |>
    group_by(subscribe) |>
    summarize(pct = n()/p_num*100) |>
    mutate(pct = round(pct, 2)) 

age_stat <- player_data |>
    summarize(average_age = mean(Age, na.rm = TRUE), min_age = min(Age, na.rm = TRUE), max_age = max(Age, na.rm = TRUE), median_age = median(Age, na.rm = TRUE)) |>
    map_dfr(~ round(.x , 2))

subscribed_pct
age_stat
gender_pct
```

The data set predominantly consists of **young males (median age is 19)**. Almost **3/4 of the players** are **subscribed** to the newsletter. 

Moving onto the `sessions.csv` file, I summarized the date coverage of the data set (first and last recorded session) as well as statistics on the duration of a session.

```r

session_duration_stats <- session_data |>
    mutate(session_duration = as.numeric(dmy_hm(end_time) - dmy_hm(start_time))/60) |>
    summarize(total_duration = sum(session_duration, na.rm = TRUE),
              avg_duration = mean(session_duration, na.rm = TRUE),
              median_duration = median(session_duration, na.rm = TRUE),
              min_duration = min(session_duration, na.rm = TRUE),
              max_duration = max(session_duration, na.rm = TRUE))


date_coverage <- session_data |>
    summarize(start_date = min(dmy_hm(start_time)),
              end_date = max(dmy_hm(end_time), na.rm = TRUE),
              duration = end_date - start_date)

date_coverage
session_duration_stats
```

Most importantly, data spans over around **173 days** and the median duration sits at a **half an hour**.

I want to address the second broad question: **We would like to know which "kinds" of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts**.

The specific question I want to explore is: **Can age, subscription status and experience level predict the number of hours played?**. Because k-nn regression only works for numerical predictors, I convert TRUE and FALSE into numeric 1 and 0 respectively, and mapping experience to scores from 1 to 5 in the following order - beginner, amateur, regular, veteran, pro.

```r
players_wrangled <- player_data |>
    mutate(experience = recode(experience,
                               "Beginner" = 1,
                               "Amateur" = 2,
                               "Regular" = 3,
                               "Veteran" = 4,
                               "Pro" = 5),
           subscribe = as.numeric(subscribe))

players_wrangled
```

Below I compute the mean values for all quantitative variables in the players dataframe. I choose to include `experience` as quantitative because the mean will give us an approximate understanding of the average experience level, which can be interpreted. `subscribe` tells us the percentage of players who subscribed to the newsletter as a decimal.

```r
player_means <- players_wrangled |>
    select(experience, played_hours, Age, subscribe) |>
    map_dfr(mean, na.rm = TRUE) |>
    round(2)

player_means
```

The mean experience value 2.71 can be interpreted as an amateur/close to regular player.

The visualization below is a scatter plot of the relationship between age and hours played on the server. Due to the dominance of younger people in the data set, it is hard to accurately assess the relationship, but based on what is available, players of age above 30 contribute less playing time, with the largest amount of hours played by individual between 15 and 25. There is no linear relationship between the two variables. 

```r
options(repr.plot.width = 8, repr.plot.height = 8)


age_vs_played_hrs <- players_wrangled |>
    ggplot(aes(x = Age, y = played_hours)) +
    scale_y_log10() +
    geom_point(alpha = 0.7, color = "steelblue") +
    labs(x = "Age of the player", y = "Hours played on the server", title = "Age vs. hours played") +
    theme(text = element_text(size = 18))
age_vs_played_hrs
```

Now let's look at a bar plor of the relationship between experience level and total hours played. It appears that Amateur (2) and Regular (3) players played the most hours on the server. Again, due to data imbalance (unequal proportions of players by level), it cannot be said with certainty that Veteran (4) and Pro (5) players contribute less playing time.

```r
options(repr.plot.width = 10, repr.plot.height = 6)

played_hrs_by_exp <- players_wrangled |>
    group_by(experience) |>
    summarize(total_hours = sum(played_hours)) |>
    ggplot(aes(x = experience, y = total_hours, fill = experience)) +
    geom_bar(stat = "identity") + 
    labs(x = "Minecraft experience level (from 1 to 5)", y = "Total hours played on the server", title = "Experience level vs. total hours played", fill = "Experience level") +
    theme(text = element_text(size = 18))

played_hrs_by_exp
```

Lastly, let's examine the relationship between newlsetter subscription status and total hours played. I split the bar plot by 5 experience levels. Subscribed players contribute more playing time across all experience levels, with the difference being largest for Amateur and Regular players. One potential explanation for this is that newsletter subscription is an indication of higher interest in the server, being a predicitve factor for engagement.

```r
options(repr.plot.width = 18, repr.plot.height = 10)

played_hrs_vs_subscription <- players_wrangled |>
    group_by(experience, subscribe) |>
    summarize(total_hours = sum(played_hours)) |>
    ggplot(aes(x = subscribe, y = total_hours, fill = experience)) +
    geom_bar(stat = "identity", width = 0.7) +
    facet_grid(cols = vars(experience)) +
    scale_x_continuous(breaks = c(0,1), labels = c("Not subscribed", "Subscribed")) +
    labs(x = "Is the player subscribed to the newsletter?", y = "Total hours played on the server", fill = "Experience level") +
    theme(text = element_text(size = 20))

played_hrs_vs_subscription
```
